# Statistical Summary

### Statistical Summary by the Length

```python
length_feature = ['SepalLengthCm', 'PetalLengthCm']

for i in length_feature:
    mean = np.mean(data[i])
    median = np.median(data[i])
    std = np.std(data[i])
    str_strip = i.strip('Cm')
    
    print(str_strip)
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Standard Deviation: {std:.2f}")
    print("-" * 30)
```

 ##### from the calcuation it shows that the average of sepal length for the species is 5.84 while the median sepal length is 5.80, and the standard deviation is dispersed not too high or too low from the mean. In the other hand, the average of petal length for the three species is 3.76, median petal length is 4.35, however, the standard deviation is dispersed slightly greater from the mean value.

### Statistical Summary by the Width

```python
width_feature = ['SepalWidthCm', 'PetalWidthCm']

for i in width_feature:
    mean = np.mean(data[i])
    median = np.median(data[i])
    std = np.std(data[i])
    str_strip = i.strip('Cm')
    
    print(str_strip)
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Standard Deviation: {std:.2f}")
    print("-" * 30)
```

##### from the calcuation it shows that the average of sepal width for the species is 3.05 while the median sepal width is 3.00, and the standard deviation is fairly dispersed from the mean value. In the other hand, the average of petal width for the three species is 1.20, median petal width is 1.30, however, the standard deviation is dispersed close around the mean value.
